## 线性回归
#### 动手学深度学习 —— 线性回归
#### 使用 Pytorch 实现——yujia

*以下内容将逐步带着实现一个**简单但是要素都全**的神经网络*   
   
如图，我们要实现一个简单的回归神经网络，其中包含两个输入$x1,x2$，对应的权重为$w1,w2$，对应的输出为$y^{'}$。举例子可以为，房屋面积和房屋地段对房屋价格的影响，因为房屋价格是连续的，所以是一个回归问题，即是通过房屋价格和房屋地段两个因素去训练对应的权重，从而实现任意房屋价格和地段对房屋价格的预测。    
$$y^{'}=w1*x1+w2*x2+b$$
这不同于西瓜问题，西瓜问题的输出是离散的，可分的——好瓜和坏瓜，因此是一个分类问题    
**【任务要求】**    
通过已知$x1,x2$的输入，和对应的$y$输出，训练尽可能贴近的权重$w1,w2$   
简单地说就是在一大堆数据中提炼输入和输出的关系

### Part1 
导入必要的库

```python
import numpy as np
from matplotlib import pyplot as plt
from IPython import display
import random
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
```

### Part2
创造一个数据集

```python
num_inputs = 2 #输入个数
num_examples = 1000 #样本总数（数据集总数）
true_w = [2, -3.4] #真实权重（仅作为造数据用，我们就是要从未知训练权重，尽可能贴近这个值）
true_b = 4.2 #真实偏执
features = np.random.normal(0, 1, (num_examples, num_inputs)) #以正态分布随机创造一些输入
features = torch.as_tensor(features) #将输入转换成张量
```

```python
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b #labels :标签，就是 y值  
labels += np.random.normal(0, 0.01, labels.shape)
labels = torch.as_tensor(labels) #转换为张量
labels = labels.resize_(num_examples,1) #将张量resize，这里后面会说，很重要
```

**可视化数据：** 画图操作，不是最重要的

```python
display.set_matplotlib_formats('svg')
plt.rcParams['figure.figsize'] = (3.5,2.5)
# plt.scatter(features[:,1], labels, 1)
```

### Part 3
**写入数据集**    
我们造的数据，或者以后从其他渠道获取的数据，需要统一处理让pytorch识别    
一般需要新建一个MyDataset类继承Dataset（Pytorch中的类）   
并重写 \_\_init__, \_\_getitem__, \_\_len__ 三个方法    
\_\_init__ : 传入数据集    
\_\_getitem__ ： 通过索引找到每个元素值    
\_\_len__ ： 数据长度    

```python
class MyDataset(Dataset):
    def __init__(self, features, labels):
        self.x = features
        self.y = labels
        self.len = len(self.x)
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    def __len__(self):
        return self.len
```

### Part 4
**读取数据**
在读取数据的时候需要用到Pytorch的Dataloader    
batch_size : 每次计算的批量，因为数据集一般很庞大，在计算时就不采用一个一个的遍历方式，通常使用将一些数据打包成一个batch来计算。一般来说，batch_size 越大，模型收敛的越快，也不绝对，可能会出现在正确值的上下波动的情况，需要具体调整    
shuffle ： 是否打乱数据集

```python
myDataset = MyDataset(features, labels)
data_iter = DataLoader(dataset=myDataset, batch_size=10, shuffle=True)
for X,y in data_iter:
    print(X,y)
    break
```

### Part 5
**设置显卡设备**    
一般在做视觉任务的时候，计算量十分庞大，需要使用显卡，这里就是通过检测torch.cuda.is_available()，来判断显卡是否准备就绪    
需要提前安装好对应版本的 CUDA　和　CUdnn

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
device
```

### Part 6
**定义神经网络**    
一般需要建一个Net类继承 nn.Module，重写\_\_init__和forward 方法       
\_\_init__ ： 理解为定义网络，骨架，这个网络都包含哪些东西，没有顺序。但是如果想要省事定义在nn.Sequential里的话就需要严格顺序要求。    
forward ： 理解为数据流向（前向传播），就是数据怎么走，如果上面定义时没有按照顺序来，则下面就要有顺序了，如果上面使用nn.Sequential定义，下面直接调用即可，会自动按照nn.Sequential声明的顺序走    

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__() #从这里往上的内容就是照抄，每次都一样的
        self.linear_net = nn.Sequential(
            nn.Linear(2,1)
        )
    def forward(self,x):
        x = self.linear_net(x)
        return x
model = Net().double()
```

**网络参数初始化**    
我们回到最原本的问题，我们现在只知道输入和输出，完全不知道权重，那要从哪里预测呢？因此就需要对权重初始化，赋一个初值，这里没有固定的方法赋值，需要自行考虑，如果初始化做的好，将大大节约训练的时间    

```python
def NetInit(m):
    if type(m) == nn.Linear:
        torch.nn.init.normal_(m.weight, mean=0, std=0.01)
model.apply(NetInit)
model.state_dict()
```

**定义损失函数**     

```python
loss = torch.nn.MSELoss()
```

**定义优化器**    
梯度下降

```python
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)
```

```python
num_epochs = 10
```

### Part 7
**训练**    
epoch ： 训练轮次    
pred ： 预测值    
**特别注意**    
如果预测值 pred 和 真实值 y 的张量结构（shape）不相同，则会引起计算错误，因此在计算损失前务必查看张量形状是否一致

```python
loss_list = []
for epoch in range(1, num_epochs+1):
    for X, y in data_iter:
        pred = model(X)  #计算预测值
        l = loss(pred, y)  #计算损失
        optimizer.zero_grad()  #每次需要先将梯度清零再反向传播
        l.backward()  #反向传播更新权重
        optimizer.step() #梯度下降
    loss_list.append(l.mean())
    print("epoch %d, loss: %lf" % (epoch, l.mean()))
```

**观察模型训练情况**    
我们发现训练的不错    
和真实值非常接近    

```python
model.state_dict()
```

```python
plt.ylabel('train_loss')
plt.plot(range(len(loss_list)),loss_list)
```
